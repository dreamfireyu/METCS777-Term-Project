```python
import pandas as pd
import glob
import os

folder_path = 'tripdata'  
file_paths = glob.glob(os.path.join(folder_path, '*.csv'))
data_list_v1 = []
data_list_v2 = []
for file_path in file_paths:
    date = file_path[9:15]
    df = pd.read_csv(file_path)
    if date <= '202303':
        data_list_v1.append(df)
    else:
        data_list_v2.append(df)

#print(file_paths)
print(len(data_list_v1))
print(len(data_list_v2))
```

```python
combined_df_v1 = pd.concat(data_list_v1, ignore_index=True)
combined_df_v1.drop_duplicates(inplace=True)

combined_df_v1.fillna({
    'postal code': 'Unknown' 
}, inplace=True)

combined_df_v1.dropna(inplace=True)
print(len(combined_df_v1))
```

```python
combined_df_v1['start station id'] = combined_df_v1['start station id'].astype(str)
combined_df_v1['end station id'] = combined_df_v1['end station id'].astype(str)
combined_df_v1['bikeid'] = combined_df_v1['bikeid'].astype(str)
combined_df_v1['usertype'] = combined_df_v1['usertype'].astype('category')
combined_df_v1.sort_values(by='starttime', ascending=True, inplace=True)

combined_df_v1[:5]
```

```python
#combined_df_v1.to_csv("BikeData_202101_202303.csv")
```

```python
combined_df_v1['Hour_start'] = combined_df_v1['starttime'].apply(lambda x: x[:13])
combined_df_v1['Hour_stop'] = combined_df_v1['stoptime'].apply(lambda x: x[:13])
combined_df_v1[:5]
```

```python
grouped_start = combined_df_v1.groupby(['start station id', 'Hour_start']).size().reset_index()
grouped_stop = combined_df_v1.groupby(['end station id', 'Hour_stop']).size().reset_index()
print(grouped_start[:-5])
print(grouped_stop[:-5])
```

```python
grouped_start = combined_df_v1.groupby(['start station id', 'Hour_start']).size().reset_index(name='Start_Count')
grouped_stop = combined_df_v1.groupby(['end station id', 'Hour_stop']).size().reset_index(name='Stop_Count')

grouped_start = grouped_start.rename(columns={'start station id': 'station_id', 'Hour_start': 'Hour'})
grouped_stop = grouped_stop.rename(columns={'end station id': 'station_id', 'Hour_stop': 'Hour'})

merged_df = grouped_start.merge(grouped_stop, on=['station_id', 'Hour'], how='outer')

merged_df = merged_df.fillna(0)

merged_df['Start_Count'] = merged_df['Start_Count'].astype(int)
merged_df['Stop_Count'] = merged_df['Stop_Count'].astype(int)
merged_df['Net_Flow'] = merged_df['Stop_Count'] - merged_df['Start_Count']
```

```python
from pandas.tseries.holiday import USFederalHolidayCalendar
calendar = USFederalHolidayCalendar()
holidays = calendar.holidays(start='2021-01-01', end='2023-04-01')
holidays = holidays.date
holidays
```

```python
merged_df['Hour'] = pd.to_datetime(merged_df['Hour'])
merged_df.set_index('Hour', inplace=True)
merged_df['hour'] = merged_df.index.hour
merged_df['day_of_week'] = merged_df.index.dayofweek
merged_df['is_holiday'] = merged_df.index.normalize().isin(holidays)
merged_df[:5]
```
