```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.tsa.ar_model import AutoReg
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Read the data
df = pd.read_csv('ar_data.csv')
df['Hour'] = pd.to_datetime(df['Hour'])
df.set_index('Hour', inplace=True)
```

```python
print(df.head())
```

```python
station_ids = df['station_id'].unique()
```

```python
results = {}
for station_id in station_ids:
    station_data = df.loc[df['station_id'] == station_id].copy()
    station_data = station_data.sort_values(by='Hour')
    
    # Split the data
    split_point = int(len(station_data) * 0.8)
    train, test = station_data['Net_Flow'][:split_point], station_data['Net_Flow'][split_point:]
    
    # Fit the model
    model = AutoReg(train, lags=3)
    model_fit = model.fit()
    
    # Predict
    pred_start = test.index[0]
    pred_end = test.index[-1]
    predictions = model_fit.predict(start=pred_start, end=pred_end)
    
    # MSE
    mse = mean_squared_error(test, predictions)
    
    # Save the model and MSE
    results[station_id] = {'Model': model_fit, 'MSE': mse}
    
    print(f"Station ID: {station_id}")
    print("Model Summary:")
    print(model_fit.summary())
    print(f"MSE: {mse}\n")
```
